```python
import os
import numpy as np
from tqdm import tqdm
import matplotlib.pyplot as plt
```

```python
path = '/home/seth/datasets/gut/stft'
```

```python
max_list = []
min_list = []

for sub_dir in tqdm(os.listdir(path)):
    for class_dir in os.listdir(os.path.join(path, sub_dir)):
        cls_path = os.path.join(path, sub_dir, class_dir)
        files = sorted(os.listdir(cls_path), key=lambda x: int(x.split('.')[0]))
        for file in files:
            X = np.load(os.path.join(cls_path, file))
            max_list.append(max(X.flatten()))
            min_list.append(min(X.flatten()))
```

```python
plt.hist(max_list, bins='auto')
plt.show()
```

```python
plt.hist(min_list, bins='auto')
plt.show()
```

```python
x = np.arange(0.00000001,20,0.01)
y = np.log10(x)
plt.scatter(x, y)
plt.show()
```

```python
plt.hist(np.log10(np.array(max_list)+1e-6)+6, bins='auto')
plt.show()
```

```python
plt.hist(np.log10(np.array(min_list)+1e-6)+6, bins='auto')
plt.show()
```
